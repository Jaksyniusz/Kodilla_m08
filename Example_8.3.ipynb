## Pivot table

```python
import pandas as pd

# Wczytaj plik Excel
df = pd.read_excel('Pivot.xlsx')

# Wyświetl pierwsze 5 wierszy
print(df.head())
```

df.head() domyślnie wyświetla 5 wierszy, aby zapewnić szybki podgląd danych.

```python
import numpy as np

# Użyj stringa 'sum' zamiast np.sum
pivot = df.pivot_table(values='Sprzedaż', index='Przedstawiciel', columns='Region', aggfunc='sum')

print(pivot)
```

```python
df.pivot_table(values='Sprzedaż',index='Przedstawiciel',columns='Region',aggfunc='sum').fillna(0).round(2)
```

```python
df.pivot_table(values='Sprzedaż',index=['Region','Przedstawiciel'],aggfunc='sum').round(2)
```

```python
df.pivot_table(values='Sprzedaż',index='Region',aggfunc=[len,'max','min']).round(0)
```

## Apply

```python
def commission_fee(x):


    if x <= 300:
        return 0
    elif x <= 900:
        return x * 0.03
    else:
        return x * 0.06
```

```python
df['commission_fee'] = df['Sprzedaż'].apply(lambda x: commission_fee(x))
```

```python
df
```

```python
df['Produkt_len'] = df['Produkt'].apply(len)
df
```

```python
import math
df['#_opakowań'] = df['Sztuki'].apply(lambda x: math.ceil(x/5))
df
```

```python
def bonus(row):

    margin = (row['Sprzedaż'] - row['Koszty'])/row['Sprzedaż']

    if margin > 0.55:
        return 200
    else:
        return 0
```

```python
df['Bonus'] = df.apply(lambda row: bonus(row),axis=1)
df.sample(10)
```
